# Exploratory Data Analysis (EDA) of Abalone Dataset

This Jupyter Notebook focuses on conducting an Exploratory Data Analysis (EDA) on the Abalone Dataset. The EDA includes distribution analysis of key variables and a correlation analysis among features, especially focusing on their relationship with the age of abalone.

## Loading the Dataset

The dataset is loaded into a pandas DataFrame. The dataset includes measurements of abalones, and we are particularly interested in predicting their age.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

abalone_data_path = 'abalone/abalone.data'
column_names = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
abalone_df = pd.read_csv(abalone_data_path, header=None, names=column_names)
abalone_df['Age'] = abalone_df['Rings'] + 1.5  # Age calculated from Rings
```

## 1. Distribution Analysis

We analyze the distributions of key variables including Length, Diameter, Height, and Age. Understanding these distributions is crucial for grasping the characteristics of the abalones in our dataset.


```python
# Setting up the plot
plt.figure(figsize=(15, 10))

# Plotting distributions of key variables
plt.subplot(2, 2, 1)
sns.histplot(abalone_df['Length'], kde=True)
plt.title('Distribution of Length')

plt.subplot(2, 2, 2)
sns.histplot(abalone_df['Diameter'], kde=True)
plt.title('Distribution of Diameter')

plt.subplot(2, 2, 3)
sns.histplot(abalone_df['Height'], kde=True)
plt.title('Distribution of Height')

plt.subplot(2, 2, 4)
sns.histplot(abalone_df['Age'], kde=True)
plt.title('Distribution of Age')

plt.tight_layout()
plt.show()
```

## 2. Correlation Analysis

Next, we explore the correlations among different features in the dataset, focusing particularly on their correlation with the 'Age' of abalone. This will help in understanding which features are most influential in predicting age.


```python
# Calculating the correlation matrix
correlation_matrix = abalone_df.iloc[:, 1:].corr()  # Excluding the 'Sex' column

# Plotting the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
